```python
# Function to mine association rules and visualize them
def mine_and_visualize_rules(data, min_support, min_confidence, min_length):
    # Convert the dataset into a one-hot encoded format
    encoded_data = pd.get_dummies(data.astype(bool))

    # Find frequent itemsets using the Apriori algorithm
    frequent_itemsets = apriori(encoded_data, min_support=min_support, use_colnames=True)

    # Generate association rules
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)

    # Filter rules based on minimum length
    rules = rules[rules['antecedents'].apply(lambda x: len(x)) >= min_length]

    # Visualize the rules using plots
    plt.figure(figsize=(10, 6))
    plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
    plt.xlabel('Support')
    plt.ylabel('Confidence')
    plt.title('Association Rules')
    plt.show()

    # Print the number of rules generated
    num_rules = len(rules)
    print(f"Support={min_support}, Confidence={min_confidence}, Min Length={min_length}, Number of Rules={num_rules}")

# Define different values for support, confidence, and minimum length
support_values = [0.05, 0.1, 0.15, 0.2]
confidence_values = [0.5, 0.6, 0.7, 0.8, 0.9]
min_length_values = [1, 2]

# Mine association rules for different parameter combinations
for support in support_values:
    for confidence in confidence_values:
        for min_length in min_length_values:
            print(f"Mining rules with support={support}, confidence={confidence}, min_length={min_length}")
            mine_and_visualize_rules(df, support, confidence, min_length)

all_rules = {}
for support in support_values:
    for conf in confidence_values:
        try:
            # Attempt to generate rules with appropriate parameters
            rules = association_rules(apriori(df.astype(bool), min_support=support, use_colnames=True),
                                       min_threshold=conf)
            all_rules.setdefault(support, {})[conf] = len(rules)
        except ValueError:  # Handle the case where no frequent itemsets are found
            print(f"No frequent itemsets found for support: {support}, confidence: {conf}")

# Create a DataFrame to visualize the number of rules at different confidence levels
all_conf_df = pd.DataFrame(all_rules).T
all_conf_df.plot(figsize=(8, 8), grid=True)
plt.ylabel('Number of Rules')
plt.xlabel('Confidence')
plt.title('Number of Rules at Different Confidence Levels (Varying Minimum Support)')
plt.show()

# Generate final rules with adjusted minimum support and confidence threshold
final_rules = association_rules(apriori(df.astype(bool), min_support=0.01, use_colnames=True),
                                min_threshold=0.4)

# Filter rules with confidence greater than 0.5 and extract support, confidence, and lift
filtered_rules = final_rules[final_rules['confidence'] > 0.5]
support = filtered_rules["support"]
confidence = filtered_rules["confidence"]
lift = filtered_rules["lift"]

# Combined visualization (3D scatter with colormap)
plt.figure()
plt.scatter(support, confidence, c=lift, cmap='gray')
plt.colorbar(label='Lift')
plt.xlabel("Support")
plt.ylabel("Confidence")
plt.title('Support vs. Confidence colored by Lift')
plt.grid(True)
plt.show()
```
